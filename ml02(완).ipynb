```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

iris = load_iris()
x, y = iris.data, iris.target
```

```python
# 결정나무 학습
dt = DecisionTreeClassifier(max_depth=2 , random_state=42)
dt.fit(x,y)

plt.figure(figsize=(12, 8))
plot_tree(dt, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
print(iris.feature_names)
print(iris.target_names)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

data = {
    '공부시간(일)': [1, 5, 8, 2, 6, 0, 9, 3, 7, 4],
    '수면시간(시)': [8, 6, 7, 4, 7, 5, 8, 3, 6, 9],
    '출석률(%)':   [60, 90, 95, 50, 85, 40, 100, 55, 90, 70],
    '합격여부':    ['불합격','합격','합격','불합격','합격','불합격','합격','불합격','합격','합격']
}

df = pd.DataFrame(data)
df.columns = ['day', 'hour', 'per', 'pass'] 
x = df[['day','hour','per']]
y = df['pass']

# 결정나무 학습
dt = DecisionTreeClassifier(max_depth=1 , random_state=42)
dt.fit(x,y)

plt.figure(figsize=(12, 8))
plot_tree(dt, feature_names= x.columns , class_names= ['불합격','합격'], filled=True, rounded=True, fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

data2 = {
    '기온(도)':   [28, 15, 10, 33, 20, 8,  25, 35, 18, 12],
    '습도(%)':    [80, 50, 40, 90, 60, 30, 55, 95, 45, 35],
    '바람(m/s)':  [3,  1,  5,  2,  4,  7,  1,  3,  6,  2],
    '운동여부':   ['안함','함','함','안함','함','안함','함','안함','안함','함']
}


df = pd.DataFrame(data2)
x = df[['기온(도)','습도(%)','바람(m/s)']]
y = df['운동여부']

# 결정나무 학습
dt = DecisionTreeClassifier(max_depth=2 , random_state=42)
dt.fit(x,y)

plt.figure(figsize=(12, 8))
plot_tree(dt, feature_names= x.columns , class_names= dt.classes_, filled=True, rounded=True, fontsize=16)
plt.show()
```
