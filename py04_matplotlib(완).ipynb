```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,100)
y = np.exp(x)
plt.plot(x,y)
plt.yscale('log') # linear
plt.show()
```

### 막대그래프

```python
# 세로막대 bar , 가로막대 barh
x = np.arange(3)
year = ['2024','2025','2026']
value =[100,400,800]
plt.bar(x, value, color=['r','g','b'])
plt.xticks(x, year)
plt.show()
```

```python
# 세로막대 bar , 가로막대 barh
x = np.arange(3)
year = ['2024','2025','2026']
value =[100,400,800]
plt.barh(x, value, color=['r','g','b'])
plt.yticks(x, year)
plt.show()
```

```python
# 막대 폭 지정
x = np.arange(3)
year = ['2024','2025','2026']
value =[100,400,800]
plt.bar(x, value, color=['r','g','b'], width=[0.3,0.5,1])
plt.xticks(x, year)
plt.show()
```

```python
# 막지 엣지 색상, 두께
x = np.arange(3)
year = ['2024','2025','2026']
value =[100,400,800]
plt.barh(x, value, color=['r','g','b'], linewidth=3, edgecolor=['pink','y','cyan'] )
plt.yticks(x, year)
plt.show()
```

```python
# 막지 엣지 색상, 두께
x = np.arange(3)
year = ['2024','2025','2026']
value =[100,400,800]
colors = ['r' if v < 400 else 'y' for v in value ]
plt.barh(x, value, color='y', linewidth=3, edgecolor=colors )
plt.yticks(x, year)
plt.show()
```

### 산점도 그래프

```python
# 산점도 기본
np.random.seed(10)
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
plt.scatter(x,y)
plt.show()
```

```python
# 산점도 크기
np.random.seed(20)
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

size = (20 * np.random.rand(n))**2
plt.scatter(x,y, s=size)
plt.show()
```

```python
# 산점도 색상
np.random.seed(20)
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

size = (20 * np.random.rand(n))**2
color = np.random.rand(n)
plt.scatter(x,y, s=size, c=color)
plt.show()
```

```python
# 산점도 투명도 컬러맵
np.random.seed(30)
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

size = (20 * np.random.rand(n))**2
color = np.random.rand(n)
plt.scatter(x,y, s=size, c=color, alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

### 파이차트 그래프

```python
# 파이차트기본
ratio = [33,15,12,40]
labels = ['Apple','Banana', 'Melon', 'Orange']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
# 파이차트 방향 지정
ratio = [33,15,22,30]
labels = ['Apple','Banana', 'Melon', 'Orange']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
# 파이차트 변형
ratio = [33,15,22,30]
labels = ['Apple','Banana', 'Melon', 'Orange']
explode = [0,0.20,0,0]
plt.pie(ratio, labels=labels, autopct='%.1f%%', explode=explode)
plt.show()
```

```python
# 파이차트 컬러 강조
ratio = [33,15,22,30]
labels = ['Apple','Banana', 'Melon', 'Orange']
colors = ['#ffff00' if v < 20 else 'whitesmoke' for v in ratio ]
explode = [0.05,0.20,0.05,0.05]
plt.pie(ratio, labels=labels, colors=colors, autopct='%.1f%%', explode=explode, shadow=True)
plt.show()
```

```python
# 파이차트 스타일주기
ratio = [33,15,22,30]
labels = ['Apple','Banana', 'Melon', 'Orange']
# colors = ['#ffff00' if v < 20 else 'whitesmoke' for v in ratio ]
colors = ['red','yellow','lightgreen','orange']
wedgeprops={'width':0.7,'edgecolor':'w','linewidth':5}
plt.pie(ratio, labels=labels, colors=colors, autopct='%.1f%%',wedgeprops=wedgeprops)
plt.show()
```

### 멀티 그래프

```python
x1 = np.linspace(0.0,5.0,50)
x2 = np.linspace(0.0,2.0,50)

y1= np.cos(2*np.pi*x1)* np.exp(-x1)
y2= np.cos(2*np.pi*x2)

plt.subplot(2, 1, 1)
plt.plot(x1,y1,'o-')
plt.title('1st Graph')

plt.subplot(2, 1, 2)
plt.plot(x2,y2,'.-')
plt.title('2nd Graph')

plt.tight_layout()
plt.show()
```
