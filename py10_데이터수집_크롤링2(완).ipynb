```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import re # 정규표현식
url = 'https://www.weather.go.kr/w/weather/forecast/short-term.do?stnId=159'
web = req.get(url)
soup = bs(web.content ,'html.parser')
summ = soup.select('.summary')
for s in summ:
    result= re.sub(r"\s+"," ",s.text) 
    print(result.replace("○","\n○"))    
```

```python
sum = soup.select('.summary')
table = soup.select('.table-col')[0]
data = table.select('tr')
header = data[0].select('th')
line1 = ['최저기온']+ data[1].select('td')
line2 = ['최고기온']+ data[2].select('td')
raw_data = []
for l0,l1,l2 in zip(header,line1,line2):
    try:
        out = l0.text+':'+l1.text+' / '+l2.text
    except:
        out = l0.text+': '+l1+' / '+l2
    finally:
        print(out)
        raw_data.append(out)
```
