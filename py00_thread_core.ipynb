```python
import time
def long_task():
    sum = 0
    for ii in range(5):
    #     time.sleep(1)
        for i in range(50*1000*1000):
            sum += i    
        print(f"일하는중...{ii}")
```

```python
# 기본작업 5초 x 5회  = 25초????
start = time.time()
print("=====시작=====")
# for n in range(5):
long_task()
print("=====끝=====")
print(time.time() - start)
```

```python
# thread를 사용함, 프로그램으로 병렬처리
import threading

start = time.time()
print("=====시작=====")

box = []
for n in range(10):
    t = threading.Thread(target=long_task)   
    box.append(t)
for th in box:
    th.start()
for th in box:
    th.join()
    
print("=====끝=====")
print(time.time() - start)
```

```python
import multiprocessing # 물리적 코어, CPU 코어를 활용해서 병렬로 작업을 수행할 수 있다.

start = time.time()
print("=====시작=====")

pro = []
for n in range(10):
    t = multiprocessing.Process(target=long_task)   
    pro.append(t)
for pr in pro:
    pr.start()
for pr in pro:
    pr.join()
    
print("=====끝=====")
print(time.time() - start)
```
