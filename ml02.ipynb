```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os
import urllib.request

# 1. 폰트 다운로드
font_url = "https://github.com/google/fonts/raw/main/ofl/nanumgothic/NanumGothic-Regular.ttf"
font_path = "./NanumGothic-Regular.ttf"

if not os.path.exists(font_path):
    print("나눔고딕 폰트를 다운로드 중입니다...")
    urllib.request.urlretrieve(font_url, font_path)
    print("다운로드 완료!")

# 2. Matplotlib에 안전하게 폰트 추가 (에러 수정 부분)
fm.fontManager.addfont(font_path)

# 3. 폰트 및 마이너스 기호 설정 적용
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

print("✅ 성공! 이번엔 진짜로 나눔고딕 폰트 설정이 완료되었습니다.")
print("이제 원래 그리려던 '결정나무 시각화 셀'을 다시 실행해 보세요!")
```

```python
# 결정나무 학습
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# 1. 데이터 로드 및 변수 정의 (대문자 X, 소문자 y 확인!)
iris = load_iris()
X = iris.data
y = iris.target

# 결정나무 학습      
dt = DecisionTreeClassifier(max_depth=2, random_state=42)

# 2. dt.fit(x, y) 대신 정의한 대문자 X를 넣어줍니다.
dt.fit(X, y)      

# 시각화
plt.figure(figsize=(12, 8))      
plot_tree(dt, feature_names=iris.feature_names, class_names=list(iris.target_names), filled=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_wine

# 1. 와인 데이터셋 로드
wine = load_wine()

# 2. 변수명을 대문자 X, 소문자 y로 통일 (관례를 따릅니다)
X, y = wine.data, wine.target

# 3. 모델 정의
dt = DecisionTreeClassifier(max_depth=2, random_state=42)

# 4. 일치된 변수(X, y)로 모델 학습 (둘 다 똑같이 178개로 매칭됨)
dt.fit(X, y)      

# 5. 시각화 (iris 대신 전부 wine의 속성으로 변경)
plt.figure(figsize=(12, 8))      
plot_tree(dt, feature_names=wine.feature_names, class_names=list(wine.target_names), filled=True)
plt.show()

# 6. 특성 이름 및 타깃 이름 확인
print("특성(Feature) 이름들:\n", wine.feature_names)
print("\n타깃(Target) 클래스 이름들:\n", wine.target_names)
```
