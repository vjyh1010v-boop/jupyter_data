```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select, WebDriverWait # 셀렉트 구문사용
from selenium.webdriver.support import expected_conditions as EC # 깨지는거 방지

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup as bs
from hdfs import InsecureClient
import pandas as pd
from datetime import datetime
import time as tt

options = Options()
# options.binary_location = './driver_u/......' # 리눅스용
# options.binary_location = r'D:\data\jupyter_data\driver_w\chrome.exe' # 크롬위치, 윈도우용, r은 raw string 로우 문자열
options.add_argument("--headless")
# options.add_argument("--no-sandbox") # windows에서는 없어도 괜찮다.
# options.add_argument("--disable-dev-shm-usage") # 공유 메모리 비활성화
# options.add_argument("--disable-gpu") # GPU 비활성화
options.add_argument("--window-size=1920,1080") # 가상 화면 크기 설정

service = Service(r'D:\data\jupyter_data\driver_w\chromedriver.exe')

driver = webdriver.Chrome(service=service, options=options )
driver.get('https://www.opinet.co.kr/searRgSelect.do')
print(driver.title)
```

```python
# 지역의 구 셀렉트창 선택후 옵션을 조회
wait = WebDriverWait(driver, 10)
select_element = wait.until(EC.presence_of_element_located((By.ID, "SIGUNGU_NM0")))
select = Select(select_element)
option_count = len(select.options)
option_count
```

```python
# 전체 옵션을 반복

try:
    for i in range(1,option_count):
        select_element = wait.until(EC.presence_of_element_located((By.ID, "SIGUNGU_NM0")))
        select = Select(select_element)
        print('순회시작:', i ,'/' , option_count -1)
        tt.sleep(2)
        select.select_by_index(i)
        print(i,'번 옵션 선택')
        selected = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "#templ_list0 > div:nth-child(8) > div > a"))) 
        selected.click()
        tt.sleep(4)
except Exception as e:
    print("문제생김, 실패...",e)
```
