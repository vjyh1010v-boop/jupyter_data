```python
import mysql.connector

try:
    conn = mysql.connector.connect(
        host="mysql",
        user="root",
        password="1234",
        database="mydb"
    )
    if conn.is_connected(): print("MySQL 서버에 성공적으로 연결되었습니다.")
except mysql.connector.Error as err:
    print("MySQL 연결 실패:", err)
```

```python
cursor = conn.cursor()
print("SQL 쿼리 준비")
```

```python
# 테이블 생성
query = """CREATE TABLE users (
    id INT,
    name VARCHAR(50),
    age INT,
    city VARCHAR(50))
"""

cursor.execute(query)
print("테이블 생성완료")
```

```python
# 테이블 확인
cursor.execute("SHOW TABLES")
tbl = cursor.fetchall()
tbl
```

```python
# 테이블 구조 확인
cursor.execute("DESC users")
desc_tbl = cursor.fetchall()
desc_tbl
```

```python
# 단일 데이터 삽입
sql = "INSERT INTO users (id, name, age, city) VALUES (%s,%s,%s,%s)"

values = (1, "Alice", 22, "Seoul")
m_values = [
    (2, "Bob", 30, "Busan"),
    (3, "Charlie", 28, "Incheon"),
    (4, "Diana", 35, "Daegu"),
    (5, "Ethan", 26, "Gwangju"),
    (6, "Fiona", 40, "Jeju"),
    (7, "George", 33, "Seoul"),
    (8, "Hana", 27, "Suwon"),
    (9, "Ian", 29, "Ulsan"),
    (10, "Jisoo", 24, "Daejeon"),
    (11, "Kevin", 31, "Changwon"),
    (12, "Lee Min", 25, "Goyang"),
    (13, "Park Soo", 32, "Pohang"),
    (14, "Choi Hye", 29, "Ansan"),
    (15, "Jung Tae", 27, "Yongin"),
    (16, "Kim Mi", 34, "Siheung")
]

cursor.execute(sql, values)
print("1 개 데이터 삽입 완료", cursor.lastrowid)

cursor.executemany(sql, m_values)
print("여러 개 데이터 삽입 완료.")
conn.commit()
```

```python
# 데이터 조회
sql = "SELECT * FROM users"
filters = " WHERE age >= 30"
cursor.execute(sql + filters)

# 방법 1:
# for n in cursor:
#     print(n)

# 방법 2
data1 = cursor.fetchall()
for d in data1:
    print(d)
```

```python
# 업데이트
sql = "UPDATE users SET age = %s WHERE name = %s"
values = (31, "Bob")
cursor.execute(sql, values)
conn.commit()
print("수정된 행 개수:", cursor.rowcount)
```
