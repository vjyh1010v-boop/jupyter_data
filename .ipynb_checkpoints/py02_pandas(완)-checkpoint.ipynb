```python
# 판다스 시리즈 딕셔너리데이터

population_dict = {
    'korea': 5182,
    'japan': 12622,
    'china': 141178,
    'usa': 32976
}
# population_dict.keys()
# population_dict.values()
# population_dict.items()

country = pd.Series(population_dict)
print(country, country.index, country.values)
print(country.ndim, country.shape, country.size)
# 시리즈 원소에 접근방법
print(country.loc['usa'])
print(country.iloc[3])
```

```python
population_dict = {
    'korea': 5182,
    'japan': 12622,
    'china': 141178,
    'usa': 32976
}
GDP_dict = {
    'korea': 169320000,
    'japan': 516700000,
    'china': 140925000,
    'usa' : 2041280000
}

pup = pd.Series(population_dict)
GDP = pd.Series(GDP_dict)
country = pd.DataFrame({"population":pup,'GDP':GDP})
country.index.name = "Country"
country.columns.name = "Info"
print(country.index, country.columns)
print(country.values, country.shape, country.size, country.ndim)
```

```python
# 과일별 가격 (원)
price = {
    "Apple": 1200,
    "Banana": 800,
    "Orange": 1500,
    "Mango": 2500
}

# 과일별 갯수
quan = {
    "Apple": 10,
    "Banana": 5,
    "Orange": 7,
    "Mango": 20}

# 위 내용을 이용하여 데이터 프레임으로 만들고 마지막컬럼은 행의 토탈가격을 붙여라.
fruit = pd.DataFrame({'price':price,'quan':quan})
fruit['won'] = fruit['price'] * fruit['quan']

# 행추가 방법1
fruit.loc['kiwi'] = [800, 11, np.nan ]
fruit.loc['kiwi','won'] = fruit.loc['kiwi','price'] * fruit.loc['kiwi','quan']

# 행추가 방법2
melon = pd.Series( data=[4500, 6, 4500 * 6], index=['price','quan','won'])
fruit.loc['melon'] = melon
fruit

# 행추가 해서 결측값 확인하기
fruit.loc['lemon'] = [200, np.nan , np.nan]
fruit.isnull() #.sum() 개수를 파악할 때

# NaN 값이 있는 행을 찾는 방법 1
fruit[fruit['won'].isnull()].index
 
# NaN 값이 있는 행을 찾는 방법 2
# fruit[ax.any(axis=1)].index

# 삭제 fruit.drop( 위에서 찾은 인덱스 넣어주면 됨.)
```

### 데이터프레임 정렬

```python
import numpy as np
import pandas as pd
print("DdataFrame: ")
data = {
    'col1':[2,1,9,8,7,4],
    'col2':['A','A','B', np.nan, 'D','C'],
    'col3':[0,1,9,4,2,3],
}
index = [1,3,2,5,0,4]
df = pd.DataFrame(data, index=index)
df
```

```python
df = df.sort_index(axis=0) # 행을 축으로 정렬
df
```

```python
df.sort_index(axis=1, ascending=False) # 열을 축으로 정렬, 내림차순
```

```python
df.sort_values('col1', ascending=False) # values 를 기준으로 정렬
```

```python
df.sort_values(['col2','col1'], ascending=[True, False]) # values 를 기준으로 이중 정렬
```

### 데이터 분석용 함수

```python
import numpy as np
import pandas as pd
data ={
    'korean': [50,60,70,30],
    'math':[30,np.nan,40,20],
}
index = ['홍길순','수퍼걸','쉬헐크','배트걸']
df = pd.DataFrame(data, index=index)
print(df.index)
print(df.columns)
df
```

```python
# count() 함수
print(df.count())
print(df.count(axis=1))
```

```python
# max() 함수
print(df.max(axis=0))
print(df.max(axis=1))
```

```python
# min() 함수
print(df.min(axis=0))
print(df.min(axis=1))
```

```python
# sum() 함수
print(df.sum(axis=0))
print(df.sum(axis=1))
```

```python
# mean() 함수
print(df.mean(axis=0))
print(df.mean(axis=1))
print(df.mean(axis=1, skipna=False))
data1 = df.mean(axis=1, skipna=False)
print(data1[~data1.isnull()])
print(data1[data1.notnull()])
```

```python
# df['math'].mean()
df['math'] = df['math'].fillna(df['math'].mean())
df
```

### 그룹으로 묶기 (groupby)

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'key'   : ['A','A','A','A','B','C'],
    'data1':[1,2,3,1,2,3,],
    'data2':[4,4,6,0,6,1],
    })
print("DataFrame:")
df
```

```python
df.groupby('key').sum()
```

```python
df.groupby(['key','data1']).sum().reset_index()
```

```python
df
```

```python
df.groupby('key').mean()
```

```python
def filter_mean(x):
    return x['data2'].mean() < 3
df.groupby('key').filter(filter_mean)
# filter_mean(df.groupby('key'))
```

```python
df.groupby('key').aggregate(['min','median','max'])
```

```python
df.groupby('key').aggregate({'data1':'mean','data2':'sum'})
```

### 피봇과 피봇 테이블

```python
import pandas as pd
df = pd.DataFrame([
    ['20210901', 'A',   1],
    ['20210901', 'B',  10],
    ['20210901', 'C', 100],
    ['20210902', 'A',   2],
    ['20210902', 'B',  20],
    ['20210902', 'C', 200],
    ['20210903', 'A',   3],
    ['20210903', 'B',  30],
    ['20210903', 'C', 300],
], columns = ['date','type', 'volume'])
df
```

```python
df.set_index('date')
```

```python
pv1 = df.pivot(index="date", columns="type", values = "volume")
pv1
```

```python
pv1_idx = pv1.reset_index()
pv1_idx
```

```python
pv1_idx.set_index('date')
```
